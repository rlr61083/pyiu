In 1911, Danish astronomer Ejnar Hertzsprung discovered that the **temperature and luminosity (brightness) of stars** are related, which was independently discovered by the American astronomer Henry Norris Russell in 1913, who plotted these quantities against each other for the first time. Named in honor of these two astronomers, the **Hertzsprung-Russell (H-R) diagram** has since become a fundamental tool in astronomy. Further research in the field has shown that, not only are temperature and luminosity related, but they are also indicative of a **star's age and evolutionary state**. Specifically, the clustering of stars in certain regions of the diagram tells us how long they spend in each phase of their life and the physics behind that evolutionary state.

<center><img src="https://raw.githubusercontent.com/pyiu-workshops/pyiu/refs/heads/main/crash-course/hr-diagram/imgs/hr_diagram.jpg" width="500" alt="Hertzsprung-Russell diagram showing where main-sequence stars, giants, supergiants, and white dwarfs lie"/></center>

As depicted in the H-R diagram above, most stars lie along a narrow sequence running from the upper left (corresponding to hot, luminous stars) to the lower right (corresponding to cooler, less luminous stars). This region is called the main sequence, which is where most stars will spend most of their lives, including our Sun. The clear correlation between temperature and luminosity along the main sequence demonstrates that hotter stars are more luminous than cooler stars.

### Magnitudes and Colors
In reality, astronomers cannot directly measure the luminosity, or intrinsic brightness, of a star. Instead, astronomers measure a star's apparent brightness, often reported in historic units called magnitudes. An important rule when dealing with magnitudes is that they are logarithmic, not linear, and that **smaller magnitudes correspond to brighter objects** while **larger magnitudes correspond to dimmer objects**. Then, astronomers combine the star's **apparent magnitude** with measurements of its distance to calculate its **absolute magnitude**, or intrinsic brightness. This is done using the famous distance modulus equation, written as

$$m-M = 5 \log(d/10)$$

where $m$ corresponds to apparent magnitude, $M$ is absolute magnitude, and $d$ is distance measured in parsecs (pc). *Star clusters*, which comprise thousands of stars densely packed together, are a key tool here, as their stars lie at roughly the same distance. Thus, knowing the distance to one star in a cluster gives us the distance to all of them!

### Stellar Evolution

<center><img src="https://raw.githubusercontent.com/pyiu-workshops/pyiu/refs/heads/main/crash-course/hr-diagram/imgs/evolutionary_tracks.webp" width="450" alt="Sample evolutionary tracks of a variety of stellar masses"/></center>

As a star evolves or ages, its luminsoity and temperature change. This causes the star to move across the H-R diagram. Once a star reaches the main-sequence stage, it begins converting hydrogen into helium through nuclear fusion in its core. This process lasts a long time, and stars spend the majority of their lives on the main-sequence. However, more massive stars use up their fuel much more quickly than low-mass stars. Eventually, once all of a star's hydrogen is used up, it moves away from the main sequence. This happens faster for more massive stars.

Astronomers make models of stellar evolution using stars of different masses and compositions to see how stars change throughout their lives. Stellar evolution models calculate how stars of different masses change over time. An evolutionary track follows a single star of fixed mass as it moves through the H-R diagram. The figure above shows evolutionary tracks for stars of different masses (0.5, 1, 1.5, 3, 5, and 15 solar masses). Each black curve traces how a single star of fixed mass moves through the H-R diagram as it ages. The numbers along each track indicate the time (in years) since the star formed. The red line is the zero-age main sequence, representing the time at which a star first enters the stable hydrogen-burning phase, which marks the main sequence.

### Star Clusters

Because stellar evolution occurs over millions to billions of years, we cannot directly observe a single star aging through all stages of its life. Instead, astronomers study star clusters, which contain many stars formed at approximately the same time. The members of these clusters are often gravitationally bound to each other, and it is assumed that they formed at the same time from the same cloud of gas. Thus, these stars are expected to have nearly the same composition, with the only variable property among them being their mass. We know that stars with larger masses evolve more quickly and will leave the main sequence to become red giants on shorter timescales. On the other hand, stars with smaller masses will remain on the main sequence for longer periods of time. The point in the H-R diagram where more massive stars begin to leave the main sequence is called the main-sequence turnoff point.

### Isochrones

<center><img src="https://raw.githubusercontent.com/pyiu-workshops/pyiu/refs/heads/main/crash-course/hr-diagram/imgs/isochrones.png" width="400" alt="Sample isochrone models of a variety of ages"/></center>

An isochrone (meaning “equal time”) represents what a group of stars of different masses would look like if they all formed at the same time and evolved for the same amount of time. While an evolutionary track follows one star of fixed mass as it changes over time, an isochrone is constructed by taking stars of many different masses and asking: "Where would these stars be in the H-R diagram at different ages?".

Because more massive stars evolve faster than lower-mass stars,
high-mass stars will already have moved off the main sequence while low-mass stars still remain on the main sequence.

This shows the main-sequence turnoff, which shifts depending on the age of the population. Each isochrone therefore corresponds to a different age.

### Icochrones: Cluster Age

To find the age of a star cluster, astronomers compare the observed H-R diagram of the stars in the cluster to the isochrones. Simply, astronomers plot different isochrones to the H-R diagram of their star cluster and find the one that matches the shape of the cluster the best. This tells us the age of the cluster!

### Icochrones: Cluster Distance

Isochrones models are given in absolute magnitudes ($M$), while the cluster data is in apparent magntiudes ($m$). So we need to perform the following calculation to get the distance to the cluster:

$$d = 10^{((m-M)/5) +1}$$

Isochrones are complex, so instead of having you create your own models, we will read in and manipulate pre-generated models in order to determine the distance and age of the cluster. Let's load in one pre-generated model and save the apparent $V$ magnitudes and $B-V$ colors to separate lists, just as we did for the actual observed data for the NGC 2420 cluster.

**Important note:** the length of the isochrone file is different from the `hr_ngc2420.txt` file since the isochrone is a model that describes the shape of the data at discrete points of interest.

```python
# Read in one pre-generated isochrone model
url_iso1 = "https://raw.githubusercontent.com/pyiu-workshops/pyiu/refs/heads/main/crash-course/hr-diagram/input-data/iso19.txt"
ngc_2420_iso1 = np.loadtxt(url_iso1, skiprows=2)

print("Length of isochrone data file (number of data points for the model):", len(ngc_2420_iso1))
print("Isochrone data start: [B-V, V_app]")
for data in ngc_2420_iso1[:5]:
  print(data)

# Create empty lists to store colors and magnitudes in
bv_colors_iso1 = []
vmags_app_iso1 = []

# Save the B-V color and apparent V magnitude of each point in the isochrone model
for i in range(len(ngc_2420_iso1)):
  bv_colors_iso1.append(ngc_2420_iso1[i][0]) # First column is B-V color
  vmags_app_iso1.append(ngc_2420_iso1[i][1]) # Second column is apparent V magnitude
```

Let's plot our isochrone model against our H-R diagram for NGC 2420.

```python
# Create the figure
plt.figure(figsize=(6, 7))
# Plot B-V colors against apparent V magnitudes
plt.scatter(bv_colors, vmags_app, s=5, c="black")
# Plot the isochrone data
plt.plot(bv_colors_iso1, vmags_app_iso1, c="red", linestyle="solid")

# Set the plot title and font size
plt.title("NGC 2420 H-R Diagram With Isochrone", size=15)
# Note: we have to invert the y-axis since smaller magnitudes are brighter!
plt.gca().invert_yaxis()

# Set the x- and y-axis labels
plt.xlabel("B-V", size=15) # B-V color
plt.ylabel("V (Apparent Magnitude)", size=15) # Apparent V magnitude

plt.show()
```

It looks like our model is pretty far off from the data! This is because this model was generated assuming a specific age for the cluster, without accounting for distance. Stars in a more distant cluster will appear fainter on the H-R diagram than stars in a nearer cluster, all other factors held equal. When we account for distance $d$, we can convert our apparent magnitudes $m$ into true absolute magnitudes $M$ using the distance modulus equation, which we used during part 1 of this H-R diagram workshop:

$$m-M = 5 \log(d/10)$$

We will need to shift our model *vertically* on the H-R diagram until it matches the data, then find the distance corresponding to that difference. This shift in magnitudes is called the **distance modulus**, which is the $m-M$ part of the equation above. After some algebra, we can solve this equation for distance as:

$$ d = 10^{((m-M)/5+1)} $$

Remember that $d$ is in units of parsecs (pc) here. Okay, so let's try shifting our isochrone downward in our H-R diagram toward *fainter* magnitudes so that it matches up with the data better. Let's try a shift of $+11$ to start.

```python
vmags_app_iso1_shifted = []
vmag_app_shift = 11 # Isochrone vertical shift

for i in range(len(ngc_2420_iso1)):
  # Apply shift to each star
  vmags_app_iso1_shifted.append(vmags_app_iso1[i] + vmag_app_shift)
```

```python
# Create the figure
plt.figure(figsize=(6, 7))
# Plot B-V colors against apparent V magnitudes
plt.scatter(bv_colors, vmags_app, s=5, c="black")
# Plot the isochrone data
plt.plot(bv_colors_iso1, vmags_app_iso1_shifted, c="red", linestyle="solid")

# Set the plot title and font size
plt.title("NGC 2420 H-R Diagram With Isochrone", size=15)
# Note: we have to invert the y-axis since smaller magnitudes are brighter!
plt.gca().invert_yaxis()

# Set the x- and y-axis labels
plt.xlabel("B-V", size = 15) # B-V color
plt.ylabel("V (Apparent Magnitude)", size = 15) # Apparent V magnitude

plt.show()
```

## **2. Logic**

What does one do if they want to identify outliers in their data, or if they'd like to categorize data based on some specific criteria? This is where the idea of logic comes into play!

Similar to `for` loops, using logic requires specific keywords and syntax to be used. Important features include:
- The `if` keyword, followed by a logical statement (usually some condition you want to test)
- Logical operators that immediately follow `if`, including the `==` operator which tests if two quantities are equal, as well as `<`, `>`, `<=`, and `>=` which test various inequalities between two quantities (less than, greater than, less than or equal to, greater than or equal to, respectively)
- Boolean operators like `and`, `or`, and `not` that can be used to combine multiple conditions in a single `if` statement
- The `else` keyword, which follows the `if` statement and automatically encapsulates all other cases
- A colon `:` at the end of each logical/conditional statement
- The operations you want to carry out if a certain condition is met. These lines of code will be automatically indented by Python to indicate they are within the conditional statement

Let's say, for example, that we had a list of numbers and we wanted to identify and flag negative values, while saving only positive values to a new list. Let's try it:

```python
# Define a list of sample numbers to work with
list1 = [-5, -3, -0.1, 0, 1, 4]
list1_positive = [] # Positive numbers in list1

# Loop through the list
for i in range(len(list1)):
  if list1[i] < 0: # Check if the element is negative
    print(list1[i], "is a negative number")
  else: # If it's not negative, we say that it's positive
    print(list1[i], "is a positive number")
    list1_positive.append(list1[i])

print("list1:", list1)
print("Positive numbers in list1:", list1_positive)
```

Something's a bit off here: 0 is not a positive number, but it was kept in our list regardless since it was not negative. For intermediate cases like this, we can make use of an additional keyword: `elif`, which has the same function as `if` but means "else if". So we can re-write our algorithm as follows:
1) Check if (`if`) a number of the list is negative, then ignore it.
2) Else if (`elif`) the number isn't negative, but ALSO isn't positive (so it is 0), we still want to ignore it.
3) Else (`else`), the number must be positive, so we add it to our positive list.

```python
# Define a list of sample numbers to work with
list1 = [-5, -3, -0.1, 0, 1, 4]
list1_positive = [] # Positive numbers in list1

# Loop through the list
for i in range(len(list1)):
  if list1[i] < 0: # Check if the element is negative
    print(list1[i], "is a negative number")
  elif list1[i] == 0: # If it's not negative, check if it's zero
    print(list1[i], "is zero")
  else: # If it's neither negative nor zero, it must be positive
    print(list1[i], "is a positive number")
    list1_positive.append(list1[i])

print("list1:", list1)
print("Positive numbers in list1:", list1_positive)
```

## **3. Functions**
Often when programming, we need to perform the same task over and over again, but with different inputs each time. To make this easier and faster, we can define a function. In the previous H-R demo, we already worked with a few built-in functions, like `print()` and `len()`, but we can also define our own using the built-in `def` and `return` keywords:

```python
# Define a function that cubes a number
def cube(n):
  return n**3

nums = list(range(1, 6)) # Generate a list of numbers from 1 to 5
cubed_nums = []
for i in range(len(nums)):
  result = cube(nums[i])  # Call our cube function with each number as an input
  cubed_nums.append(result)

print("Original numbers:", nums)
print("Cubed numbers:", cubed_nums)
```

Python functions can perform more complicated tasks than simple mathematical functions, too! Let's say we want to separate all the odd and even numbers from a list of integers into two distinct lists. We can use the modulus operator (`%`) in Python, which returns the remainder after dividing by a number, to do this. Let's design our function to return a nested list containing an "even" and "odd" list that we can separate.

```python
# Split a list of integers into a list of even integers and a list of odd integers
def split_even_odd(lst):
  # Create empty lists to store the even and odd integers
  even = []
  odd = []
  # Loop through every number in the input list
  for num in lst:
    # If a given number is even, append it to the "even" list
    if num % 2 == 0:
      even.append(num)
    # If it's odd instead, append the number to the "odd" list
    elif num % 2 != 0:
      odd.append(num)
  # Return both new lists in one nested list
  return [even, odd]

even_odd = [62, 3, 91, 24, 77, 390, 500, 201, 462]
even_odd_lists = split_even_odd(even_odd)
even = even_odd_lists[0]
odd = even_odd_lists[1]
print("Even numbers in list:", even)
print("Odd numbers in list:", odd)
```

### Application to the H-R Diagram

Let's apply what we've learned about functions to the H-R Diagram. We'll start by writing a function to calculate isochrones at different vertical (magnitude) offsets, which each correspond to different distances (i.e., different distance modulus values).

Similar to the even/odd example above, we'll want to create a nested `list` containing several lists of shifted isochrone models at different distance modulii. In this case, we should use a `for` loop to build the nested list, and simply define our function to return a single shifted isochrone given some apparent magnitude offset.

```python
# Shift each element in the given isochrone list by the same distance_modulus value
def shift_isochrone(vmags_isochrone, distance_modulus):
  shifted_isochrone = []
  for i in range(len(vmags_isochrone)):
    shifted_isochrone.append(vmags_isochrone[i] + distance_modulus)
  return shifted_isochrone

# Define offsets (distance moduli) of interest
vmag_offsets = [11, 13, 15]
# Plug offsets into function
shifted_isochrones = []
for i in range(len(vmag_offsets)):
  shifted_isochrones.append(shift_isochrone(vmags_app_iso1, vmag_offsets[i]))

print("Distance moduli tested:", vmag_offsets)
```

Now, let's plot these shifted isochrone models to see which one fits the data best. Play around with the `vmag_offsets` list to try and get the most precise fit!

```python
# Create the figure
plt.figure(figsize=(6, 7))
# Plot B-V colors against apparent V magnitudes
plt.scatter(bv_colors, vmags_app, s=5, c="black")

# Define a color list
iso_colors = ["blueviolet", "green", "blue"]
# Plot the isochrones
for i in range(len(shifted_isochrones)):
  plt.plot(bv_colors_iso1, shifted_isochrones[i], c=iso_colors[i],
           linestyle="solid", label=f"$m-M={vmag_offsets[i]}$")

# Set the plot title and font size
plt.title("NGC 2420 H-R Diagram With Isochrones", size=15)
# Note: we have to invert the y-axis since smaller magnitudes are brighter!
plt.gca().invert_yaxis()

# Set the x- and y-axis labels
plt.xlabel("B-V", size=15) # B-V color
plt.ylabel("V (Apparent Magnitude)", size=15) # Apparent V magnitude

# Add a legend to show different offsets tested
plt.legend(loc="lower left")

plt.show()
```

You should find that a distance modulus of about $m-M\approx +12$ magnitudes provides the closest alignment to the cluster data. Use this best-fit offset to estimate the distance to this cluster (recall $d = 10^{((m-M)/5+1)} $).

```python
vmag_offset_best = 12

d_pc = 10**(vmag_offset_best/5+1)

print("The distance to NGC 2420 is", d_pc, "pc.")
# Round the result to the nearest whole number using the built-in "round" function
print("The distance to NGC 2420 is", round(d_pc), "pc.")
```

We now have three isochrone models stored in our code, all shifted to the correct distance. These isochrones correspond to ages of 1.9 Gyr, 2.2 Gyr, and 2.5 Gyr, respectively. Our last step is to plot all three of them on top of our cluster data and see which isochrone is the best fit to the data!

```python
# Create the figure
plt.figure(figsize=(6, 7))
# Plot B-V colors against apparent V magnitudes
plt.scatter(bv_colors, vmags_app, s=5, c="black")

# Define a color list
iso_colors_age = ["darkmagenta", "forestgreen", "royalblue"]
# Write out the ages of the different models (these will go in the legend)
ages = ['1.9 Gyr', '2.2 Gyr', '2.5 Gyr']
# Plot the isochrones
for i in range(len(shifted_isochrones_age)):
  plt.plot(bv_colors_iso_all[i], shifted_isochrones_age[i], c=iso_colors_age[i],
           linestyle="solid", label=f"Age = {ages[i]}")

# Set the plot title and font size
plt.title("NGC 2420 H-R Diagram With Isochrones", size=15)
# Note: we have to invert the y-axis since smaller magnitudes are brighter!
plt.gca().invert_yaxis()

# Set the x- and y-axis labels
plt.xlabel("B-V", size=15) # B-V color
plt.ylabel("V (Apparent Magnitude)", size=15) # Apparent V magnitude

# Restrict the y-axis limits to focus on the relevant region of the plot
plt.ylim(18.2, 10.5)

# Add a legend to show ages of different isochrone models
plt.legend(loc="upper left")

plt.show()
```

Which isochrone matches the data best? How old is this cluster?

Looking at the isochrones plotted above, we can see that the isochrone corresponding to an age of **2.2 Gyr** best matches the data. The 1.9 Gyr isochrone is a bit too blue around the main-sequence turnoff (in other words, too far to the left), and the 2.5 Gyr isochrone is a bit too red (too far to the right). Therefore, we can conclude that **NGC 2420 is about 2.2 Gyr old.**

## Further Considerations

There's much more we can learn from isochrones and H-R diagrams of star clusters!

1) In science, we're often interested in the uncertainty of a given measurement, which quantifies how confident we are in its value. For example, if we report the age of a star cluster as 3 &plusmn; 2 Gyr, that means we're reasonably confident its age is somewhere between 1 and 5 Gyr. However, if we report the cluster's age as 3.0 &plusmn; 0.1 Gyr, that has a much smaller uncertainty (0.1 Gyr as opposed to 2 Gyr), making it clear that we're much more confident that the age of the cluster is around 3 Gyr.

    We can estimate an uncertainty on the age of NGC 2420 of 0.15 Gyr by examining the plot with the three different isochrones. The isochrones are separated in age by 0.3 Gyr, and we can tell that the data points fall closer to the 2.2 Gyr isochrone than to either of the other two. Can you think of a more quantitative way we could obtain an uncertainty of our age estimate?
