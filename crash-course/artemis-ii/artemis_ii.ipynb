Python has many built-in variable *types*. Four simple variable types are as follows:

- ```int```: An integer number. Unlike many other coding languages, an integer in python can be of any length. Integers are initialized by integer-valued numbers (e.g., ```ex_variable = 1```).
- ```float```: A decimal number. For very large or very small floats, often abbreviated into the form 'XeY', equivalent to $X * 10^Y$ for numbers X and Y. Floats are initialized by decimal-valued numbers (e.g., ```ex_variable = 1.05```).
- ```bool```: A boolean, or a true/false value. Booleans are initialized by True or False. (e.g., ```ex_variable = False```).
- ```str```: A string, or a sequence of text characters. Strings are initialized by quotation marks surrounding a sequence of characters. (e.g., ```ex_variable = "False"```).

This explains why ```print("Hello World!")``` above simply output ```Hello World!```---the quotation marks indicated that the characters contained within were to be read as a string-type variable, and thus the value of that string was output.

To change the type of a variable (which is called `casting` it to another type), we use a command of the same name as the type to which the variable is being cast. For example, to cast the number 1 to a string "1", one would write ```str(1)```. The type of a variable can be found with the command ```type()```.

As can be seen in the example above, this can be done implicitly (without defining a variable for the number input into the command). Note also that the ```print()``` command  casts all inputs to strings before outputting.
