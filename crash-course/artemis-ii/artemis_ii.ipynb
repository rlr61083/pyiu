This information is stored in a **variable**. Variables are defined given a name (which in Python must be a sequence of alphanumeric characters and underscores beginning with a letter or underscore) and are given a value by following that name with the `=` operator.

### 1.2 Types

Python has many built-in variable **types**. Four simple variable types are as follows:

- `int`: An integer number. Unlike many other coding languages, an integer in Python can be of any length. Integers are initialized as whole numbers (e.g., `ex_variable = 1`).
- `float`: A decimal number. For very large or very small floats, they're often abbreviated into the form `XeY`, equivalent to $X \times 10^Y$ for numbers `X` and `Y`. Floats are initialized as decimal-valued numbers (e.g., `ex_variable = 1.05`).
- `bool`: A boolean, or a true/false value. Booleans are initialized as `True` or `False` (e.g., `ex_variable = False`).
- `str`: A string, or a sequence of text characters. Strings are initialized as a sequence of characters surrounded by quotation marks (e.g., `ex_variable = "False"` or `ex_variable = 'True'`).

This explains why `print("Hello World!")` above simply output `Hello World!` &mdash; the quotation marks indicated that the characters contained within were to be read as a string-type variable, and thus the value of that string was output. ### 1.3 Operations

Python also has built-in support for simple operations between variables. This includes mathematical operations (applicable to both integers and floats), string operations (applicable to strings) and boolean operations (applicable to all types). To apply an operation between variables, simply write `a (operator) b`, where `a` and `b` are the variables or variable names in question.

Python supports the four classical mathematical operations: `+`, `-`, `*`, and `/`. Mathematical operations also include the modulus operator `%` (where `a%b` outputs the remainder of the division operation $a/b$), and the exponentiation operator `**` (remark: the `^` operator also exists, but is *not* an exponentiation operator).

Lastly, boolean operations compare the values of two variables. Boolean operators always output a boolean. They consist of the operators `==`, which returns `True` if two variables are equal and `False` otherwise, and `!=`, which returns `True` if two variables are not equal and `False` otherwise.

The boolean operators `>`, `<`, `>=`, and `<=` also exist, but they work only on variables whose values can be compared (typically just numerical variables).
